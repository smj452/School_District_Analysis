```python
# Add our dependencies
import pandas as pd
import os
```

```python
file_to_load = os.path.join("Resources", "students_complete.csv")
students_df = pd.read_csv(file_to_load)
students_df
```

```python
# put the student names in a list
student_names = students_df["student_name"].tolist()
student_names
```

```python
students_to_fix = []
for name in student_names:
    # print(name.split(), len(name.split()))
    if len(name.split()) >= 3:
        students_to_fix.append(name)
print(len(students_to_fix))
```

```python
print(students_to_fix)
```

```python
prefixes = []
for name in students_to_fix:
    if len(name.split()[0]) <= 4:
        prefixes.append(name.split()[0])
print(prefixes)
```

```python
suffixes = []
for name in students_to_fix:
    if len(name.split()[-1]) <= 3:
        suffixes.append(name.split()[-1])
        
print(suffixes)
```

```python
# Get the unique items in the "prefixes" list.
set(prefixes)
```

```python
# Get the unique items in the "suffixes" list.
set(suffixes)
```

```python
# Using strip
for name in students_to_fix:
    print(name.replace("Dr. ", ""))
```

```python
prefixes_suffixes = ["Mr. ", "Mrs. ","Dr. ","Miss ", "Ms. ","MD","PhD", "DVM", "DDS"]
for psfix in prefixes_suffixes:
    students_df["student_name"] = students_df["student_name"].str.replace(psfix, "")
students_df.head(10)
```

```python
# Put the cleaned students' names in another list.
student_names = students_df["student_name"].tolist()
student_names
```

```python
student_to_fix = []
for name in student_names:
    if len(name.split()) >= 3:
        student_to_fix.append(name)
print(len(student_to_fix))
```
